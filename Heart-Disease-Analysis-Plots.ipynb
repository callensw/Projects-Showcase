# Heart Disease Analysis
The goal of this project is to plot heart disease analysis data from 'heart-disease.csv'
* I will use the Object Oriented plotting method to visualize the data

### Step 1: Getting Started

```python
# Imports

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load data
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# Select patients ages 50 or older
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')
```

### Step 2: Creating the plot using OO Method

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
        
# Change ax0 x axis limits
ax0.set_xlim([50, 80])    

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');


# Add data to ax1
scatter_2 = ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"],
                        cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
        
# Change ax1 axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter_2.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
